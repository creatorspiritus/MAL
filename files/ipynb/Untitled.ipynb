# Планы реализации проекта МАЛ

Планы реализации проекта развития региональных воздушных перевозок с применением самолётов российского производства могут иметь своё начало из любого опорного аэродрома проекта. Количество опорных аэродромов проекта может варьироваться от 36 до 51 подразделения. 

36 опорных аэродромов - минимальное количество распределённых по всей территории Российской Федерации обособленных подразделений, расстояние между которыми составляет не более 1000 км по воздушным трассам.

51 опорный аэродром - максимальное количество опорных аэродромов, при котором распределение обособленных подразделений распределены по территории Российской Федерации оптимально, а руководство аэропортов и/или муниципальных властей подтвердили свою готовность участвовать в реализации проекта.

Базовые варианты развития проекта:

- центральный (начало из аэропорта Новокузнецк);
- южный (начало реализации проекта из аэропорта Сухум, Республика Абхазия);
- восточный (начало из аэропорта Магадан).

Все три базовых варианта в период реализации первого (основного) этапа проекта формируют "Южный Воздушный Коридор" - опорные аэродромы вдоль южных границ Российской Федерации. Данные подразделения обеспечат интенсивную эксплуатацию и обслуживание расчётных типов проекта, а также смогут предоставлять услуги по обслуживанию новой российской авиатехники иностранным авиакомпаниям ЕврАзЭС.

Для каждого опорного аэродрома определяется направление развития в соответствии со стратегическим планом реализации.

```python
# Базовый план № 1
# Аэропорт начала реализации проекта: UNWW (Новокузнецк)
# Количество опорных аэродромов проекта: 51
# Количество опорных аэродромов I этапа (ЮВК): 17
БОА = "UNWW"
```
